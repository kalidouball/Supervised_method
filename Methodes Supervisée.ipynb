# Methode 1 : KNN

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import math
from statistics import mode
```

```python
# Calcul distance entre 2 points
def get_distance(p,q):
    d=np.sqrt(sum((p-q)**2))
    return d
```

```python
# Calcul les distances entre un point et un ensemble des points
def get_distances(p,X):
    D=[get_distance(p,q) for q in X]
    return D
```

```python
# Fonction qui determine les k plus proches voisins
def k_plus_proches(k,D):
    # D : Liste des distances entre p et les autres
    index_proches=np.argsort(D)[:k]
    return index_proches
```

```python
def prediction_moi(k,p,X,y):
    # p : Represente la personne
    # X : Ensemble des personnes
    # Y : Liste de ce que tes amis sont
    D=get_distances(p,X) # Recupere les distances entre p et tous les autres
    plus_proches=k_plus_proches(k,D) # Recupere les index des k plus proches
    preds=y[plus_proches] # Ce que tes plus proches sont
    vote=mode(preds) # Ce que tu es
    return vote
```

```python
def prediction (k,X,Y):
    preds=[prediction_moi(k,X[x,:],X,Y) for x in range (X.shape[0])]
    return preds
```
